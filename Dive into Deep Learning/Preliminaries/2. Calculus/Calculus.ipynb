## Derivatives and Differentiations

#### To illustrate derivatives, let us experiment with an example. Define  u=f(x)=3x<sup>2</sup>âˆ’4x .

```python
%matplotlib inline
import numpy as np
from IPython import display
from d2l import tensorflow as d2l
```

```python
# function f(x)
def f(x):
    return 3*x**2 - 4*x
```

```python
# numerical limit
def numerical_limit(f, x, h):
    return (f(x + h) - f(x)) / h
```

```python
# limit calculation
h = 0.1
x = 1
for i in range(5):
    print(f'h = {h: .5f}, numerical limit = {numerical_limit(f, x, h): .5f}')
    h *= 0.1
```

```python
# derivative visualization
x = np.arange(0, 3, 0.1)
x
```

```python
f(x)
```

```python
d2l.plot(x, [5*x**2 - 3*x, 7*x - 5], 'x', 'f(x)', legend=['f(x)', 'Tangent line (x=1)'])
```

## Partial Derivatives

```python
d2l.plot(x, [x**3 - (1/x), 4*x - 4], 'x', 'f(x)', legend=['f(x)', 'Tangent line (x=1)'])
```
